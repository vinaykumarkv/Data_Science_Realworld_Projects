```python
def fib(n):
    if n<=2:
        result=1
    else:
        result = fib(n-1) + fib(n-2)
    return result

print(fib(40))
```

```python
def fib_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(n - 1):
        a, b = b, a + b
    return b

print(fib_iterative(100))
```
