# Pharma Patient Experience & Satisfaction Prediction

# Business Understanding 
## Business Problem
Pharma teams receive thousands of patient reviews/ratings from online communities.  
Early detection of dissatisfied patients can help:
- Quality & manufacturing → investigate potential batch issues
- Medical affairs → monitor real-world safety/effectiveness signals
- Sales & marketing → target adherence support programs to at-risk patients

## Primary Modeling Goal
Predict whether a patient is likely to be **dissatisfied** (rating ≤ 4 out of 10) based on review text + available metadata.

## Success Criteria
- High **recall** on dissatisfied class (catch most unhappy patients) → target ≥ 80–85% recall
- Acceptable precision (not too many false alarms) → target ≥ 60–70%
- Primary metric: **F1-score** (balance of precision & recall)
- Secondary: ROC-AUC, confusion matrix, precision-recall curve

## Baseline to beat
1. Dummy classifier (always predict majority class = "satisfied") → ~85–90% accuracy but 0% recall on dissatisfied
2. Simple rule: predict dissatisfied if review contains negative keywords (we'll build this as a quick baseline)
3. Mean rating baseline (regression view): predict average rating (~7–8) → high RMSE on low ratings

## Stretch Goal 
Build a simple lookup/recommendation table:  
For each condition → top 3 drugs with highest average rating + enough reviews
